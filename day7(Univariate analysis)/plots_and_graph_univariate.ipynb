```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
df.info()
```

```python
df.head()
```

# Categorical Data

```python
#countplot
sns.countplot(data=df, x='Sex', hue="Sex")
```

```python
sns.countplot(data=df,x='Embarked',hue='Embarked')
```

```python
#pie chart
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# barplot
df['Embarked'].value_counts().plot(kind='bar')
```

# numerical 

```python
# histogram
plt.hist(df['Survived'])
```

```python
plt.hist(df['Pclass'],bins=3)
```

```python
# displot
sns.displot(df['Age'])
```

```python
# distplot
sns.distplot(df['Age'])
```

```python
# boxplo
sns.boxplot(data=df,x='Age',orient='h',hue='Age')
```
